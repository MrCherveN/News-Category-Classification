```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#2. Порівняння моделей

На цьому етапі виконано порівняння результатів трьох моделей, побудованих у попередніх ноутбуках:

* Baseline (TF-IDF + Logistic Regression)
* TF-IDF + SVD + XGBoost
* SentenceTransformer + Logistic Regression

Порівняння виконується за результатами на валідаційному наборі даних.

```python
results = pd.DataFrame({
    "Model": [
        "Baseline (TF-IDF + Logistic Regression)",
        "TF-IDF + SVD + XGBoost",
        "SentenceTransformer + Logistic Regression"
    ],
    "Accuracy": [0.5921, 0.4726, 0.6118],
    "Macro F1": [0.4345, 0.2960, 0.4763],
    "Weighted F1": [0.5634, 0.4322, 0.5911],
    "Час побудови моделі": ["00:02:29", "01:21:12", "00:03:27"]
})

results
```

```python
plt.figure(figsize=(8,5))

plt.bar(results["Model"], results["Accuracy"])

plt.ylabel("Accuracy")
plt.title("Порівняння моделей за Accuracy")

plt.xticks(rotation=15)

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.bar(results["Model"], results["Macro F1"])

plt.ylabel("Macro F1")
plt.title("Порівняння моделей за Macro F1")

plt.xticks(rotation=15)

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.bar(results["Model"], results["Weighted F1"])

plt.ylabel("Weighted F1")
plt.title("Порівняння моделей за Weighted F1")

plt.xticks(rotation=15)

plt.show()
```

***Висновок:***
* Порівняння трьох моделей показало, що найкращі результати на валідаційному наборі даних продемонструвала модель SentenceTransformer + Logistic Regression.

* Порівняно з базовою моделлю (TF-IDF + Logistic Regression) використання SentenceTransformer дозволило покращити всі основні метрики якості класифікації.

* Модель TF-IDF + SVD + XGBoost показала найнижчі результати серед протестованих моделей, що свідчить про недостатню ефективність цього підходу для даної задачі.

* За результатами порівняння саме SentenceTransformer + Logistic Regression було обрано як фінальну модель для проведення оцінювання на тестовому наборі даних.
